Para esta práctica se ha seleccionado una base de datos proviniente de Kaggle llamada [Chicago House Price](https://www.kaggle.com/datasets/tawfikelmetwally/chicago-house-price?resource=download).

Esta es una base de datos que nos muestra las carácteristicas principales de las casas dentro de los suburbios de Chicago.

La basa de datos contiene las siguientes variables:

* Price : Precio de la casa.
* Bedroom: Numero de cuartos.
* Room: Número de recámaras.
* Space : Tamaño de la casa (en pies cuadrados).
* Lot : Anchura del lote.
* Tax : Monto de impuestos anuales.
* Bathroom : Número de baños.
* Garage : Número de garage.
* Condition: Condición de la casa (1 si es bueno , 0 otro).


## Estructura de los Datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("realest.csv")
df.head(10)
```

```python
df.info()
```

Podemos observar como nuestra base de datos cuenta con valores nulos.
Por lo que procederemos a quitarlos.

```python
df1 = df.dropna()
df1.head(10)
```

```python
df1.info()
```

```python
df1.dtypes
df1.shape
```

Podemos observar que ahora nuestra base de datos paso a ser de 157 a 128 entradas con 9 variables.

```python
df1.describe()
```

```python
r=5
c=2
it=1
for i in df1:
    plt.subplot(r,c,it)
    sns.kdeplot(x = df1.loc[:,i], data = df1)
    plt.grid()
    it+=1
plt.tight_layout()
plt.show()
     
```

De lo anterior podemos observar que el numero del precio de vivienda probablemente siga una distribución normal. A su vez, notamos que el impuesto puede llegar a seguir una distribución logarítmica normal con tendencia hacia los 800 dólares. En cuestión de los baños por casa vemos una tendencia fuerte entre 1 y 2 baños por unidad. Por último podemos deducir que la mayor parte de las casas no se consideran en un buen estado.

```python
matriz_correlacion = df1.corr()
```

```python
sns.heatmap(matriz_correlacion, annot=True, cmap='RdYlGn')
plt.title('Matriz de Correlación')
plt.show()
```

De la matriz anterior podemos ver que el precio y el espacio tienen una relación lineal alta y positiva. A su vez, prácticamente todas las varibles tienden a tener una relación lineal positiva con el precio, a excepción de la condición de la casa que tiene una relación lineal nula - negativa (por lo que esperas una buena casa pagando más por ella).
Por otro lado, también observamos que a mayor espacio más cuartos, recámaras, impuestos y garage tiene la casa.

```python
sns.regplot(x = df1['Price'], y = df1['Space'], data = df1)
```

Por último, podemos ver en esta imágen justo lo mencionado en el punto anterior, a como aumenta el precio aumenta a su vez el espacio de la vivienda.
